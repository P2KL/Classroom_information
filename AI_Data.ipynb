```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the data
data = pd.read_csv('sales_data_cleaned.csv')
# Summarize the total sales by customer
customer_sales = data.groupby('ชื่อลูกค้า')['ยอดขาย'].sum().sort_values(ascending=False)
# Take the top customers for better visibility in the chart
top_customers = customer_sales.head(10)
# Create a vertical bar chart for the top customers' sales
plt.figure(figsize=(12, 8))
top_customers.plot(kind='bar', color='skyblue')
plt.title('Top Customer Sales')
plt.xlabel('Customer Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
# Filter the data for USA and other countries
usa_data = data[data['ประเทศ'] == 'USA']
other_countries_data = data[data['ประเทศ'] != 'USA']
# Summarize the orders by product type for USA and other countries
usa_summary = usa_data.groupby('ประเภทสินค้า')['จำนวนที่สั่งซื้อ'].sum()
other_countries_summary = other_countries_data.groupby('ประเภทสินค้า')['จำนวนที่สั่งซื้อ'].sum()
# Combine the data into a single DataFrame
combined_summary = pd.DataFrame({'USA': usa_summary, 'Other Countries': other_countries_summary}).fillna(0)
# Create a 100% stacked column chart
ax = combined_summary.div(combined_summary.sum(axis=1), axis=0).plot(kind='bar', stacked=True, figsize=(10, 6))
ax.set_title('100% Stacked Column Chart of Product Orders by Country')
ax.set_xlabel('Product Type')
ax.set_ylabel('Percentage of Orders')
ax.legend(title='Country Group')
# Show the chart
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the data
data = pd.read_csv('sales_data_cleaned.csv')
# Filter data for 2003 and 2004 only
data_filtered = data[data['ปี'].isin([2003, 2004])]
# Group data by year, country, and summarize orders
summary_by_year_country = data_filtered.groupby(['ปี', 'ประเทศ'])['จำนวนที่สั่งซื้อ'].sum().unstack()
# Initialize a DataFrame to store USA and Other Countries data
final_summary = pd.DataFrame()
# Sum USA data
final_summary['USA'] = summary_by_year_country['USA']
# Sum other countries' data by subtracting USA data from the total for each year
final_summary['Other Countries'] = summary_by_year_country.fillna(0).sum(axis=1) - final_summary['USA'].fillna(0)
# Plot the comparison
final_summary.plot(kind='bar', figsize=(10, 6))
plt.title('Comparison of Orders Between 2003 and 2004 for USA and Other Countries')
plt.xlabel('Year')
plt.ylabel('Total Orders')
plt.xticks(rotation=0)
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = 'path_to_your_file/clean.csv'
data = pd.read_csv('sales_data_cleaned.csv')

# Convert 'วันที่สั่งซื้อ' to datetime
data['วันที่สั่งซื้อ'] = pd.to_datetime(data['วันที่สั่งซื้อ'], errors='coerce')

# Filter data for the years 2003 and 2004
filtered_data = data[data['ปี'].isin([2003, 2004])]

# Group by country and calculate the total sales
country_sales = filtered_data.groupby(['ประเทศ', 'ปี'])['ยอดขาย'].sum().reset_index()

# Find the country with the highest sales for each year
max_sales_2003 = country_sales[country_sales['ปี'] == 2003].sort_values(by='ยอดขาย', ascending=False).iloc[0]
max_sales_2004 = country_sales[country_sales['ปี'] == 2004].sort_values(by='ยอดขาย', ascending=False).iloc[0]

# Create a DataFrame with the results
top_countries = pd.DataFrame([max_sales_2003, max_sales_2004])

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(top_countries['ปี'].astype(str) + ' - ' + top_countries['ประเทศ'], top_countries['ยอดขาย'], color=['blue', 'green'])
plt.xlabel('Country and Year')
plt.ylabel('Total Sales')
plt.title('Top Countries by Sales in 2003 and 2004')
plt.show()
```
